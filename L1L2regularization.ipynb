```
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from google.colab import files

uploaded = files.upload()
```

```

import io

data = pd.read_csv(io.BytesIO(uploaded['MELBOURNE_HOUSE_PRICES_LESS.csv']))
data.head()
```

```
data.shape
data = data.drop(['Address'], axis=1)
data.columns
data.isna().sum()
```

```
# cleaning our data
data['Price'] = data['Price'].fillna(data.Price.mean())

data.isna().sum()
```

```
# handling categorical data
data = pd.get_dummies(data, drop_first=True)
data.head()
```

```
# preparing train test split

X = data.drop('Price', axis=1)
Y = data['Price']

from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(X,Y,test_size = 0.3, random_state = 2)
```

```
# applying regular linear regression model

from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_x,train_y)

reg.score(test_x,test_y)
```

```
# regularizing through lasso

from sklearn import linear_model

lasso_reg = linear_model.Lasso(alpha=50,max_iter=100,tol=.1)

lasso_reg.fit(train_x,train_y)

lasso_reg.score(test_x,test_y)
```

```
# regularizing through Ridge

from sklearn import linear_model

Ridge_reg = linear_model.Ridge(alpha=50,max_iter=100,tol=.1)

Ridge_reg.fit(train_x,train_y)

Ridge_reg.score(test_x,test_y)
```

```
# the L1 and L2 regularization significantly increased the prediction and decreased the overfitting of model
```
